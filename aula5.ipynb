```python
top_10 = df.sort_values(by='Total', ascending=False).head(10)
```

```python
top_10
```

```python
import matplotlib.pyplot as plt
```

```python
sns.set_theme()
```

```python
ax = sns.barplot(data=top_10, y=top_10.index, x='Total',orient='h')
ax.set(
    title='Os dez países com mais imigrantes para Canadá',
    xlabel='Total de Imigrantes',
    ylabel=''
)
plt.show
```

```python
def gerar_grafico_paleta(pallete):
  fig_, ax = plt.subplots(figsize=(8,4))
  ax = sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h', palette=pallete)
  ax.set_title('Os dez países com mais imigrantes para Canadá',
               loc='left',
               fontsize=16)
  ax.set_xlabel('Total de Imigrantes', fontsize=14)
  ax.set_ylabel('')
  plt.show()
```

# Gráficos Interativos com Plotly

```python
import numpy as np
```

```python
import plotly.express as px
```

```python
anos = list(map(str,range(1980,2014)))
```

```python
brasil = df.loc['Brasil',anos]
```

```python
brasil_dict = {'ano':brasil.index.tolist(),'imigrantes':brasil.values.tolist()}
```

```python
dados_brasil = pd.DataFrame(brasil_dict)
```

```python
dados_brasil.head()
```

```python
fig = px.line(dados_brasil,
              x='ano',
              y='imigrantes',
              title='Imigrantes do Brasil para o Canadá\n de 1980 a 2013')
# centralizar o título geral
fig.update_layout(title_x=0.5)
fig.update_layout(width=1000,
                  height=500,
                  xaxis={'tickangle':-45})
fig.update_traces(line_color='#9751FC',line_width=6)
#fig.update_xaxes(tickangle=-45)
fig.show()
```

```python
# Salvando Gráficos
fig.write_html('imigrantes_brasil.html')
```
