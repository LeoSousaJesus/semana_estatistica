### Criando um dicionário e montando um dicionário personalizado

```python
brasil_dict
```

```python
df_brasil = pd.DataFrame(brasil_dict)
```

## Criando o gráfico de linhas

```python
plt.plot(df_brasil['ano'],df_brasil['imigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010','2013'])
plt.yticks([500,1000,1500,2000,2500,3000])
plt.grid()
plt.show()
```

# Transformando um gráfico em figura
- aumenta o potencial dos comandos

```python
# 8 - polegadas
# 4 - polegadas
# # 1 polegada = 2.54 cm
# largura 8 = 20.32 cm
# altura  4 = 10.16 cm
# DPI (dots per inch, ou pontos por polegada).
# Padrão no matplotib - DPI = 100
# Largura : 8 polegadas x 100 DPI = 800 pixels.
# Altura  : 4 polegadas x 100 DPI = 400 pixels.
plt.figure(figsize=(8,4))
plt.plot(df_brasil['ano'],df_brasil['imigrantes'])
plt.title("Imigração do Brasil para o Canadá")
plt.xlabel('Anos')
plt.ylabel('Número de imigrantes')
plt.xticks(['1980','1985','1990','1995','2000','2005','2010','2013'])
plt.grid()

plt.show
```

### Criando Subplots e adicionando rótulos

```python

# Criando subplots

# Criando a figura
fig,ax = plt.subplots(figsize=(8,4))

# Sublot: Gráfico de linha
ax.plot(df_brasil['ano'], df_brasil['imigrantes'], lw=2, color='#5E4CE0')
ax.set_title('Imigração so Brasil para o Canadá \n 1980 a 2013')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_xlabel('Ano', fontsize=14)
ax.set_ylabel('Número de Imigrantes', fontsize=14)
plt.show()
```

### Gráfico de linha e boxplot em subplots

```python
fig , axs = plt.subplots(1,2, figsize=(8,4))
# subplot 1 - gráfico de linha
axs[0].plot(df_brasil['ano'],df_brasil['imigrantes'])
axs[0].set_title('Imigração do Brasil para o Canadá \n 1980 a 2013')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Número de Imigrantes')
axs[0].grid()

# subplot 2 - gráfico boxplot 
axs[1].boxplot(df_brasil['imigrantes'])
axs[1].set_title("Boxplot da imigração do Brasil para o canadá")
axs[1].set_xlabel("Brasil")
axs[1].set_ylabel("Número de imigrantes")
plt.show()
```

## Comparativo dos maiores países da América Latina (Imigração Canadá)

```python
# Criar uma figura
fig, axs = plt.subplots(2,2,figsize=(10,6))

# Brasil
axs[0,0].plot(df.loc['Brasil',anos])
axs[0,0].set_title('Brasil')

# Colômbia

axs[0,1].plot(df.loc['Colômbia',anos])
axs[0,1].set_title('Colômbia')

# Argentina

axs[1,0].plot(df.loc['Argentina',anos])
axs[1,0].set_title('Argentina')

# Peru
axs[1,1].plot(df.loc['Peru',anos])
axs[1,1].set_title('Peru')

fig.subplots_adjust(hspace=0.5,wspace=0.3)

for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))

for ax in axs.flat:
    ax.set_xlabel('Ano')
    ax.set_ylabel('Números de Imigrantes')

# Usando ravel() - achatar a matriz evitando o uso complexo do for
# Usando set_ylim() - definir o limite dos valores

# definir o limite mínimo
ymin = 0

# definir o limite máximo
ymax = 7000

# criar um for usar ravel() - achatar uma matriz bidimensional em unidimensional
for ax in axs.ravel(): # uso do array unidimensional
    # usar set_ylim() - determinar o mínimo e o máximo do eixo y
    ax.set_ylim(ymin,ymax)

# suptitle
fig.suptitle("Imigração na AMérica Latina",fontsize=16)
# Customizando o Matplotlib (cores, fontes, bordas)
```

## Gráfico de Barras

```python
cores = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown', 'slateblue', 'gray', 'olive', 'navy', 'teal', 'tomato']
```

```python
fig,ax = plt.subplots(figsize=(12,5))
ax.bar(america_sul.index, america_sul['Total'],color=cores)
ax.set_title('Imigração da América do Sul para o Canadá')
ax.set_xlabel('')
ax.set_ylabel('Número de Imigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
plt.show
```

## Deixando o gráfico na horizontal

- Criar um gráfico invertido
- De vertical para horizontal

```python
fig,ax = plt.subplots(figsize=(12,5))
ax.barh(america_sul.index, america_sul['Total'],color=cores)
ax.set_title('Imigração da América do Sul para o Canadá')
ax.set_xlabel('')
ax.set_ylabel('Número de Imigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
plt.show
```

```python
america_sul_sorted = america_sul.sort_values(by='Total',ascending=True) 
```

```python
america_sul_sorted.head()
```

```python
fig,ax = plt.subplots(figsize=(12,5))
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'],color=cores)
ax.set_title('Imigração da América do Sul para o Canadá')
ax.set_xlabel('')
ax.set_ylabel('Número de Imigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
plt.show
```

# Seu chefe pediu para que o gráfico tenha um destaque para o Brasil
## O Brasil deve ficar na cor verde
## E os outros com a cor prata

```python
#Algoritmo de percorrer e colorir
# Passo 1: zerar as cores, definindo uma variável "cores"

cores = []

# Passo 2 criar uma estrutura dos paises em relação ao dataframe América do Sul

for pais in america_sul_sorted.index:

    # Passo 3 definir a cor usada para o país Brasil
    if pais == 'Brasil':
        cores.append('green')

    # Passo 4 definir a cor usado pelos demais países    
    else:
        cores.append('silver')


# criando uma variável fig para uma figura
# ax  é a variável  que representa o gráfico (eixo x)
fig, ax = plt.subplots(figsize=(12, 5))


ax.barh(america_sul_sorted.index, america_sul_sorted['Total'],color=cores)

ax.set_title('Imigração da América do Sul para o Canadá\n1980 a 2013',fontsize=18)

ax.set_xlabel('Número de Imigrantes')

# Mudar as fontes dos ticks
ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_tick_params(labelsize=12)

plt.show()
```

## O chefe pediu para que a barra mostrem os valores

```python
# i = índice
# v = valor

for i,v in enumerate(america_sul_sorted['Total']):
    print(f"O valor de i: {i}, o valor de v: {v}")
```

## Melhorando o gráfico

```python
#Algoritmo de percorrer e colorir
# Passo 1: zerar as cores, definindo uma variável "cores"

cores = []

# Passo 2 criar uma estrutura dos paises em relação ao dataframe América do Sul

for pais in america_sul_sorted.index:

    # Passo 3 definir a cor usada para o país Brasil
    if pais == 'Brasil':
        cores.append('green')

    # Passo 4 definir a cor usado pelos demais países    
    else:
        cores.append('silver')


# criando uma variável fig para uma figura
# ax  é a variável  que representa o gráfico (eixo x)
fig, ax = plt.subplots(figsize=(12, 5))


ax.barh(america_sul_sorted.index, america_sul_sorted['Total'],color=cores)

ax.set_title('Imigração da América do Sul para o Canadá\n1980 a 2013',fontsize=18)

ax.set_xlabel('Número de Imigrantes')

# Mudar as fontes dos ticks
ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_tick_params(labelsize=12)

# i = índice
# v = valor

for i,v in enumerate(america_sul_sorted['Total']):
    #print(f"O valor de i: {i}, o valor de v: {v}")
    ax.text(v,i,str(v),color='darkgreen',fontsize=10)
ax.set_frame_on(False)
ax.get_xaxis().set_visible(False)
plt.show()
```
